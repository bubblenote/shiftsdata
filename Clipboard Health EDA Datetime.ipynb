```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shifts_df = pd.read_csv('shifts.csv')
```

```python
hcp_df = pd.read_csv('hcps.csv')
```

```python
#run code only once or else it will keep adding 2019
shifts_df['created_at'] = shifts_df['created_at'] + '/2019'
shifts_df['shift_start'] = shifts_df['shift_start'] + '/2019'
```

```python
shifts_df['created_at'] = pd.to_datetime(shifts_df['created_at'], format = '%m/%d/%Y')
shifts_df['shift_start'] = pd.to_datetime(shifts_df['shift_start'], format = '%m/%d/%Y')
shifts_df
```

```python
shifts_df.info()
```

```python
shifts_df['Days to Fill Shift'] = abs((shifts_df['created_at'] - shifts_df['shift_start'])/np.timedelta64(1, 'D'))
shifts_df.head(20)
```

```python
#Isolate for those non-cancelled, verified shifts
verified_shifts_df = shifts_df[(shifts_df['is_cancelled'] == False) & (shifts_df['is_verified'] == True)]
verified_shifts_df.head(20)
```

```python
instant_fill = verified_shifts_df[verified_shifts_df['Days to Fill Shift'] == 0.0]
instant_fill
```

```python
hcp_to_hcf = verified_shifts_df.groupby('hcp_id')['hcf_id'].nunique().to_frame()
hcp_to_hcf
```

```python
#It's different frequencies than the previous ones, because it only includes verified, non-cancelled shifts 
#190 workers here vs 200 in the other frequency graph
#10 workers had all their shifts completely cancelled 
hcp_to_hcf['hcf_id'].value_counts()
```
